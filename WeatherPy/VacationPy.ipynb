```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import gmaps
import scipy.stats as st
from scipy.stats import linregress
import datetime
from API_Keys import weather_api_key
from API_Keys import googlemaps_api_key


df_weather = pd.read_csv("c:\\Users\\aleja\\onedrive\\AIO Python\\Bootcamp_Projects\\python-api-challenge\\weatherPy\\cities.csv")

# Configure gmaps
gmaps.configure(api_key=googlemaps_api_key)



# Store latitude and longitude in locations
locations = df_weather[["Latitude", "Longitude"]]

# Store Humidity in humidity
humidity = df_weather["Humidity"]

# Plot Heatmap
fig = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
max_intensity = np.max(humidity)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, dissipating=False, max_intensity=100, point_radius=3)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Narrow down the cities with wind speed less than 10 mph, cloudiness equals to 0 and max temp between 60 and 80
vacation_options = df_weather.loc[ (df_weather["Cloudiness"] < 20) &
                                  (df_weather["Humidity"]<50) &
                                   (df_weather["Max Temp"] > 25) & (df_weather["Max Temp"] < 35)].dropna()

vacation_options
```

```python
# create hotel_df with hotel name column
hotel_df = vacation_options
hotel_df["Hotel Name"] = np.nan
hotel_df.reset_index()
```

```python
# Look here for more information - https://developers.google.com/maps/documentation/places/web-service/search
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {"name" : "hotel",
          "radius" : 5000,
          "key" : googlemaps_api_key,}
```

```python
# Iterate through citiies in the hotel_df dataframe
for index,city in hotel_df.iterrows():
    
    # set the location parameter to be the city co-ordinates
    params["location"] = f"{city['Latitude']},{city['Longitude']}"

    # Call googlemaps API for a hotel closest to these co-ordinates

    nearesthotel = requests.get(base_url, params=params).json()['results']
    
    # If hotel found then add the hotel to the list otherwise skip
    try:
        print(f"  Nearest hotel in {city['City']} : {nearesthotel[0]['name']}.")
        hotel_df.loc[index, "Hotel Name"] = nearesthotel[0]['name']
    # Display something to tell us that there is no hotel within 5000m
    except Exception as e:
        print("  *Error*", city['City'], " does not have a hotel located within 5000m")
```

```python
# Using the template add the hotel marks to the heatmap
plotdatabox = """
<dl>
<dt>Hotel</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [plotdatabox.format(**city) for index, city in hotel_df.iterrows()]
hotel_locations = hotel_df[["Latitude", "Longitude"]]
hotel_plots = gmaps.marker_layer(hotel_locations, info_box_content = hotel_info)

# Add the layer to the map
fig.add_layer(hotel_plots)

# Display Map
fig
```
